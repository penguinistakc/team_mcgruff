```python
import pandas as pd
```

```python
raw_df = pd.read_csv('raw_data/Moon_Data_2017.csv')
```

```python
raw_df.head()
```

```python
raw_df.isnull().sum()
```

```python
# drop all rows with null values, create work_df
work_df = raw_df.dropna(axis=0, how='all')
```

```python
work_df.count()
```

```python
work_df.isnull().sum()
```

```python
#replace 'Waxing Cresent' and 'New moon' with correct spellings / formats
work_df['Moon Phase'] = work_df['Moon Phase'].replace(['Waxing Cresent'], 'Waxing Crescent')\
.replace(['New moon'], 'New Moon')
```

```python
work_df.head()
```

```python
#convert moon date to datetime, CCYY-MM-DD format
work_df['Date']= pd.to_datetime(work_df['Date'])
```

```python
work_df.head()
```

```python
#save new csv in clean_data folder
work_df.to_csv('clean_data/Moon_Data_2017_cleaned.csv', index=False)
```
